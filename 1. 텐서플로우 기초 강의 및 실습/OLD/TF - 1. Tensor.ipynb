# Tensor

텐서는 [이름, 크기 데이터형] 으로 이뤄져 있다

텐서를 만드는 방법은 다양하다. 지금은 두 가지 방법만 사용해서  D=A*B+C 식을 만들어 본다.
1. tf.constant
2. tf.Variable

```
import tensorflow as tf


A = tf.constant(value=[1,2,3], name='A', shape=[3], dtype=tf.int32)
B = tf.constant(value=[5,6,7])
C = tf.constant([2,5,6])
print(A)
print(B)
print(C)

D=A*B+C

print(D.name, D.shape, D.dtype)

sess = tf.Session()
result = sess.run(D) #(1*5+2=7, 2*6+5=17, 3*7+6=27)
print(result)
```

```
A = tf.Variable(initial_value=[1,2,3], name='A', dtype=tf.int32)
B = tf.Variable(initial_value=[5,6,7])
C = tf.Variable([2,5,6])
print(A)
print(B)
print(C)

D=A*B+C

print(D.name, D.shape, D.dtype)

sess = tf.Session()
sess.run(tf.global_variables_initializer()) # Variable을 사용할 땐 꼭 tf.global_variables_initializer() 을 돌려야합니다!
result = sess.run(D) #(1*5+2=7, 2*6+5=17, 3*7+6=27)
print(result)
```
