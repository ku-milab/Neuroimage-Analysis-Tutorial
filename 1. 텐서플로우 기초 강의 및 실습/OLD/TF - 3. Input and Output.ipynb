```
import tensorflow as tf

w = tf.constant([0.5, 0.7])
X = tf.placeholder(shape=(2), dtype=tf.float32)

y = w*X

with tf.Session() as sess: # 코드 상 이 블록을 벗어나면, Session은 자동으로 종료된다
    fetch_dict = {"y": y}
    feed_dict = {X: [10, 70]}
    result = sess.run(fetch_dict, feed_dict)
# Session 종료!

print(result["y"])
```

```
import numpy as np
with tf.Session() as sess:
    for step in range(5):
        # X에 무작위의 정수를 집어넣어본다
        x1 = np.random.randint(low=0, high=5)
        x2 = np.random.randint(low=0, high=5)
        feed_dict.update({X: [x1, x2]})
        result = sess.run(fetch_dict, feed_dict)
        print("Step %d"%step, "\ty1: %.1f y2: %.1f"%(*result["y"],), "\tx1: %d x2: %d"%(x1, x2))
```
