## 선형 회기 (Linear Regression)

$$
\begin{align}
y&=wx
\\
y^{target}&=2x
\\\\
\epsilon &= y^*-y
\\
w_{new}&=w+\lambda\epsilon
\\
y^{new}&=w_{new}x
\end{align}
$$

```
import tensorflow as tf
import numpy as np

epoch_cnt = 5
step_cnt = 10000
learning_rate = 0.00001

# y=w*X을 만들되, w 는 Variable로 만든다.
w = tf.Variable(0.0001, dtype=tf.float32, name='w')
X = tf.constant([1,2,3,4,5,6,7,8,9], dtype=tf.float32, name='X')

y_predict = w*X


# 목표: y=2*x, 즉 w=2가 되길 원함!
y_target = 2 * tf.random_uniform(shape=X.shape)


# 예측값과 목표값의 차이를 가중치 w에 더해준다
error = tf.reduce_mean(y_target-y_predict)
w_new = tf.assign(w, w + learning_rate*error)


sess = tf.Session()
sess.run(tf.global_variables_initializer())
trn_cnt = 0
error_list = np.empty(shape=(epoch_cnt*step_cnt), dtype=np.float32)
for epoch in range(epoch_cnt):
    for step in range(step_cnt):
        _, current_error = sess.run([w_new, error])
        error_list[trn_cnt] = current_error
        if trn_cnt%5000 ==0 :
            print("Epoch %d"%epoch, "Step %d"%step, "Error %f"%current_error)
        trn_cnt +=1
        
# W 를 tf.Variable --> tf.constant 으로 바꾸면? Error!!
```

```
import matplotlib.pyplot as plt
plt.plot(error_list)

# 시각화의 중요성!
```

###tf.Variable(name)
항상 새로운 variable을 생성

같은 이름이 있을 경우 name 뒤에 번호를 붙임

###tf.get_variable(name)
사전에 name으로 정의 된 variable이 있을 시 불러오고 (variable_scope 에서 reuse=True)

없을 시 새로운 variable을 생성

### tf.variable_scope(name, reuse)
여러 variable을 그룹 짓고 싶을 때

또는, 이미 사용 한 variable을 재사용 할 때


```
a = tf.Variable(0.0001, dtype=tf.float32, name='a')
b = tf.Variable(0.0002, dtype=tf.float32, name='a')
print(a, b)
```

```
# 에러 발생!!!
c = tf.get_variable(name='b', shape=1, dtype=tf.float32)
d = tf.get_variable(name='b', shape=1, dtype=tf.float32)
print(c, d)
```

```
with tf.variable_scope("A", reuse=tf.AUTO_REUSE): # tf.AUTO_REUSE --> 처음 생성 시 False, 이미 있을 경우 True
    e = tf.get_variable(name='c', shape=1, dtype=tf.float32, initializer=tf.random_normal_initializer())

with tf.variable_scope("A", reuse=tf.AUTO_REUSE):
        d = tf.get_variable(name='c', shape=1, dtype=tf.float32)

    
print(e, d)

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    print(sess.run(e))
    print(sess.run(d))
```
