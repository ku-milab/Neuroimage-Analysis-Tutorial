```
import tensorflow as tf
import numpy as np

import matplotlib.pyplot as plt


from tensorflow.examples.tutorials.mnist.input_data import read_data_sets as load_mnist
mnist = load_mnist("MNIST_data/", one_hot=True)

train_image = mnist.train.images.reshape(-1, 28, 28, 1)
test_image = mnist.test.images.reshape(-1, 28, 28, 1)

train_label = np.array(mnist.train.labels, dtype=np.bool)
test_label = np.array(mnist.test.labels, dtype=np.bool)
```

```
fig, axes = plt.subplots(1, 10)
for cnt, ax in enumerate(axes):
    ax.imshow(train_image[train_label[:,cnt]][0,...,0])
    ax.axis("off")

print("MNIST 이미지 크기: %d x %d"%(*train_image.shape[1:-1],))
print("학습 데이터 개수: %d"%len(train_image))
print("실험 데이터 개수: %d"%len(test_image))
```

```
X = tf.placeholder(shape=(None, 28, 28, 1), dtype=tf.float32) # 배치 개수는 정의되지 않아도 됩니다. 즉, 매 학습 스탭마다 배치 개수가 달라도 됩니다.
out_channels = 1
```

```
# NN API
w = tf.Variable(tf.random_normal(shape=[3, 3, 1, out_channels])) #[filter_height, filter_width, in_channels, out_channels]
b = tf.Variable(tf.random_normal(shape=[out_channels]))

y= tf.nn.conv2d(input=X,
                filter=w,
                strides=[1, 1, 1, 1],
                padding="VALID")

y = tf.nn.bias_add(y, b)

y = tf.nn.sigmoid(y)

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    result = sess.run(y, feed_dict={X: train_image[0:1]})

fig, axes = plt.subplots(1, 2)

axes[0].imshow(np.squeeze(result))
axes[1].imshow(train_image[0,...,0])
plt.show()

```

```
# Layers API
y = tf.layers.conv2d(inputs= X,
                     filters= out_channels,
                     kernel_size= 3,
                     activation=tf.nn.sigmoid,
                     kernel_initializer=tf.random_normal_initializer())


with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    result = sess.run(y, feed_dict={X: train_image[0:1]})

fig, axes = plt.subplots(1, 2)

axes[0].imshow(np.squeeze(result))
axes[1].imshow(train_image[0,...,0])
plt.show()
```
